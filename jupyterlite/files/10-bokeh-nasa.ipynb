# Show viewer

```python
# ///////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
def MyApp(doc):
    
    # THERE IS NOT OPENVISUS SERVER so it's not supported by the py backend
    assert(os.environ.get("VISUS_BACKEND","cpp")=="cpp")

    # profile=sealstorage
    os.environ["AWS_ACCESS_KEY_ID"]="any"
    os.environ["AWS_SECRET_ACCESS_KEY"]="any"
    os.environ["AWS_ENDPOINT_URL"]="https://maritime.sealstorage.io/api/v0/s3"

    # multiple urls
    urls=[f"https://maritime.sealstorage.io/api/v0/s3/utah/nasa/dyamond/idx_arco/face{zone}/u_face_{zone}_depth_52_time_0_10269.idx?cached=1" for zone in range(6)]
    palette,palette_range="Turbo256",(-30,60)
    field=None
    logic_to_pixel=[(0.0,1.0), (0.0,1.0), (0.0,10.0)]

    # urls.field=["https://maritime.sealstorage.io/api/v0/s3/utah/nasa/dyamond/mit_output/llc2160_arco/visus.idx?cached=1"],None
    # urls,field=["http://atlantis.sci.utah.edu/mod_visus?dataset=cmip6_cm2&cached=idx"],"ssp585_tasmax"

    view=Slices(
        show_options=["num_views","palette","dataset","timestep","timestep-delta","field","viewdep","quality","num_refinements","play-button", "play-sec"],
        slice_show_options=["direction","offset","viewdep","status_bar"])

    view.setNumberOfViews(3)
    view.setDatasets([(url,str(I)) for I,url in enumerate(urls)],"Zone")
    view.setDataset(urls[0])
    view.setQuality(-3)
    view.setNumberOfRefinements(3)
    view.setPalette(palette) 
    view.setPaletteRange(palette_range)
    view.setTimestepDelta(10)
    view.setField(field)
    view.setLogicToPixel(logic_to_pixel)
    view.setDirections([('0','Long'),('1','Lat'),('2','Depth')])
    doc.add_root(view.getBokehLayout(doc=doc))
    
ShowBokehApp(MyApp)
```
